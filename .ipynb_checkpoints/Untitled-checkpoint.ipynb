# Train across all data

```python
import pandas as pd
import os
from glob import glob

# Adjust path if needed
folder_path = 'simulated-data-raw/data/'  # or absolute path
all_files = sorted(glob(os.path.join(folder_path, '*.pkl')))

df_all = pd.concat([pd.read_pickle(file) for file in all_files], ignore_index=True)

print(f"Loaded {len(df_all)} rows.")
print(df_all.columns)
```

```python
df_all.rename(columns={
    'TX_AMOUNT': 'amount',
    'CUSTOMER_ID': 'from_id',
    'TERMINAL_ID': 'to_id',
    'TX_TIME_SECONDS': 'time',     # or 'TX_TIME_DAYS'
    'TX_FRAUD': 'label'
}, inplace=True)
```

```python
df_all['from_id'] = df_all['from_id'].astype('category').cat.codes
df_all['to_id'] = df_all['to_id'].astype('category').cat.codes

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_all[['amount', 'time']] = scaler.fit_transform(df_all[['amount', 'time']])
```

```python
# from sklearn.model_selection import train_test_split
# from sklearn.ensemble import RandomForestClassifier
# from sklearn.metrics import classification_report

# X = df_all[['from_id', 'to_id', 'amount', 'time']]
# y = df_all['label']

# X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

# # Train with class_weight to handle imbalance
# clf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
# clf.fit(X_train, y_train)

# # Evaluate
# y_pred = clf.predict(X_test)
# print(classification_report(y_test, y_pred, zero_division=1))

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, precision_recall_curve
import matplotlib.pyplot as plt
import numpy as np

# Prepare data
X = df_all[['from_id', 'to_id', 'amount', 'time']]
y = df_all['label']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, random_state=42
)

# Train with class_weight to handle imbalance
clf = RandomForestClassifier(
    n_estimators=100,
    class_weight='balanced',
    random_state=42
)
clf.fit(X_train, y_train)

# Predict probabilities
y_proba = clf.predict_proba(X_test)[:, 1]

# Set a lower threshold to improve recall
threshold = 0.1  # You can tweak this value
y_pred = (y_proba >= threshold).astype(int)

# Evaluation
print(classification_report(y_test, y_pred, zero_division=1))

# Plot Precision-Recall curve
precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
plt.plot(recall, precision, label="PR Curve")
plt.axvline(x=recall[np.argmax(precision >= 0.8)], color='r', linestyle='--', label='Recall @ Precâ‰¥0.8')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.grid()
plt.show()
```

```python
import joblib

# Save the model to a file
joblib.dump(clf, 'fraud_model.pkl')
```

```python
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(df_all['from_id'].astype('category').cat.categories, 'from_id_categories.pkl')
joblib.dump(df_all['to_id'].astype('category').cat.categories, 'to_id_categories.pkl')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assume df_all has columns: 'time' and 'label'
plt.figure(figsize=(12, 6))

# Plot time distributions by label
sns.kdeplot(data=df_all[df_all['label'] == 0], x='time', label='Non-Fraud', fill=True, common_norm=False)
sns.kdeplot(data=df_all[df_all['label'] == 1], x='time', label='Fraud', fill=True, common_norm=False)

plt.title("Distribution of Transaction Times by Fraud Label")
plt.xlabel("Time")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
